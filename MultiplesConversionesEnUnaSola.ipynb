```python
lista_de_listas = list()
lista_de_listas.append(lista_valores_serie_cabecera)
for i in range(1,266): 
    lista_valores_series_siguiente = df['columnaunica'][i].split(';') 
    lista_de_listas.append(lista_valores_series_siguiente)
```
