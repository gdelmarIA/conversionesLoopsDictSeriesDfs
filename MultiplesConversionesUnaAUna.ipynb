```python
cabeceras = ['Country Name;Country Code;1960;1961;1962;1963;1964;1965;1966;1967;1968;1969;1970;1971;1972;1973;1974;1975;1976;1977;1978;1979;1980;1981;1982;1983;1984;1985;1986;1987;1988;1989;1990;1991;1992;1993;1994;1995;1996;1997;1998;1999;2000;2001;2002;2003;2004;2005;2006;2007;2008;2009;2010;2011;2012;2013;2014;2015;2016;2017;2018;2019;2020;2021']
```

```python
df_cab = pd.DataFrame({'columnaunica':cabeceras}, index =[0])
```

```python
df_cab
```

```python
# aÃ±ado una fila al principio, que sean las cabeceras
df = pd.concat([df_cab, df]).reset_index(drop = True)
df
```

```python
# Cada fila, la hago un dataframe, cuyas cabeceras, seran en todos los data frame iguales
```

```python
lista_cabecera = list()
valores_cabecera = df['columnaunica'][0].split(';')
```

```python
longitud = len(valores_cabecera)
```

```python
for x in range(longitud):
    lista_cabecera.append(valores_cabecera[x])
```

```python
serie_cabecera = pd.Series(lista_cabecera)
```

```python
# Ya tengo la serie_cabacera que servirÃ¡ como cabecera de cada dataframe

# Hago ahora la serie primera
```

```python
lista_1 = list()
valores_1 = df['columnaunica'][1].split(';')
for x in range(longitud):
    lista_1.append(valores_1[x])
serie_1 = pd.Series(lista_1)
```

```python
# Lo uno, cabecera y serie_1 como un dataframe
```

```python
# Primero, hago un diccionario
dict_1 = {serie_cabecera[i]: serie_1[i] for i in range(longitud)}

# Y ahora paso ese diccionario a dataframe
df_sinT_1 = pd.DataFrame.from_dict(dict_1, orient='index')

# Y para que quede cabecera en cabecera y fila en fila, traspongo
df1 = df_sinT_1.T
df1
```

```python
# Hago lo mismo para otra fila original
lista_2 = list()
valores_2 = df['columnaunica'][2].split(';')
for x in range(longitud):
    lista_2.append(valores_2[x])

serie_2 = pd.Series(lista_2)

# la serie_cabecera, ya la hice una vez, no hace falta hacerla mas veces
dict_2 = {serie_cabecera[i]: serie_2[i] for i in range(longitud)}

df_sinT_2 = pd.DataFrame.from_dict(dict_2, orient='index')

df2 = df_sinT_2.T
df2
```

```python
# Uno estos dos datasets en uno solo
df_ = pd.concat([df1, df2])
df_
```
